```python
f = open("tst.dat","r")
res = np.loadtxt(f)
```

```python
param = res[:,0:2]
species = ['n','p','D','T','He3','He4','Li6','Li7','Be7']
mean = res[:,2:11]
high = res[:,11:20]
low = res[:,20:29]
```

```python
Yp_obs=0.245
Yp_err=0.003
D_obs=2.547*10**-5
Derr=0.035*10**-5
He3_obs=0.83
He3_err=0.15
```

```python
x=np.unique(param.T[0])
y=np.unique(param.T[1])
X, Y = np.meshgrid(x,y)
D_sim=mean[:,2]
H_sim=mean[:,1]
DH_sim=D_sim/H_sim
Z=DH_sim.reshape((len(x),len(y))).T
```

```python
plt.contourf(X,Y,Z,levels=[D_obs+5*Derr,1])
plt.contourf(X,Y,Z,levels=[0,D_obs-5*Derr],colors='red')
plt.xscale('log')
plt.yscale('log')
plt.show()
```
